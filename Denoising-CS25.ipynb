```python
# MRI  1
img_noisyg_mri = random_noise(img_mri, var=0.1,mode='gaussian')
img_noisyg_mri2 = random_noise(img_mri, var=0.25,mode='gaussian')
img_noisyg_mri3 = random_noise(img_mri, var=0.5,mode='gaussian')

img_noisysp_mri = random_noise(img_mri, mode='s&p', salt_vs_pepper=0.5, amount = 0.1)
img_noisysp_mri2 = random_noise(img_mri, mode='s&p', salt_vs_pepper=0.5, amount = 0.25)
img_noisysp_mri3 = random_noise(img_mri, mode='s&p', salt_vs_pepper=0.5, amount = 0.5)
```
